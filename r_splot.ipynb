```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000)
```

```python
d_r =2.5* t
d_s =3 *(t-5)
```

```python
fig, ax= plt.subplots()
plt.title("Bank Robber Caught")
plt.xlabel("Time")
plt.ylabel("Distance")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r,c="red")
ax.plot(t,d_s, c="blue")
plt.axvline(x=30, color ='purple',linestyle ='--')
plt.axhline(y=75, color ='purple',linestyle ='--')
```
